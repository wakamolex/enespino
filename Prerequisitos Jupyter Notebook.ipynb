#### Palabras mágicas

Una palabra mágica es un comando especial en Jupyter para que le permite hacer interacción rápida en el notebook de algunas situaciones especiales.  Por ejemplo cambiar directorio, trabajar iteractivamente con matplotlib.

Podemos diferenciar estas palabras mágicas porque delante de ellas están el % o %%.

Por ejemplo... podemos medir el tiempo que tarda un código en ejecutarse.

````python
def foo():
    x = 0
    for i in range(10000000):
        x = x + 1
````

````python
%timeit foo()
````

Para saber que tanto demora un segmento de código en ejecutarse lo podemos hacer por lo siguiente

````python
%%timeit

for i in range(10):
    for j in range(100):
        pass
````

#### Embeber visualizaciones de un notebook

Como vimos anterioremente podemos crear gráficos y embeberlos en el notebook por medio de un comando mágico, veamos nuevamente un ejemplo

````python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt
import numpy as np
````

````python
fig = plt.figure()
ax = fig.gca(projection='3d')

# Plot a sin curve using the x and y axes.
x = np.linspace(0, 1, 100)
y = np.sin(x * 2 * np.pi) / 2 + 0.5
ax.plot(x, y, zs=0, zdir='z', label='curve in (x, y)')

# Plot scatterplot data (20 2D points per colour) on the x and z axes.
colors = ('r', 'g', 'b', 'k')

# Fixing random state for reproducibility
np.random.seed(19680801)

x = np.random.sample(20 * len(colors))
y = np.random.sample(20 * len(colors))
c_list = []
for c in colors:
    c_list.extend([c] * 20)
# By using zdir='y', the y value of these points is fixed to the zs value 0
# and the (x, y) points are plotted on the x and z axes.
ax.scatter(x, y, zs=0, zdir='y', c=c_list, label='points in (x, z)')

# Make legend, set axes limits and labels
ax.legend()
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Customize the view angle so it's easier to see that the scatter points lie
# on the plane y=0
ax.view_init(elev=20., azim=-35)

plt.show()
````

#### Creando Slideshows

El slideshow es bastante dinámico, se crea por medio de View\Cell Toolbar\Slideshow y luego se elige cual de cada una de las celdas es slideshow.
<img src="imgs/edit_slideshow.gif" width="850"/>

Seguido tenemos que correr el siguiente comando en la consola en la carpeta donde se encuentra el notebook para convertirlo en HTML.
